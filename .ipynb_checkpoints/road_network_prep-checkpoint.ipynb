### Imports and constants

```python
import math
import numpy as np
import pandas as pd
import shapely
import matplotlib
from matplotlib import pylab as plt
import geopandas as gpd
import pyproj
import contextily as ctx

from shapely.geometry import Point, LineString, MultiPoint, Polygon, GeometryCollection
from shapely.ops import unary_union
from shapely.wkt import loads as wkt_load
```

```python
crs_gps = pyproj.CRS.from_epsg('4326') # WGS84 Merkator (degrees)
crs_web = pyproj.CRS.from_epsg('3857') # Pseudo-Merkator (meters)
crs_est = pyproj.CRS.from_epsg('3301') # Estonian Coordinate System
```

### Data import and exploration

Dataset downloaded from: https://geohub.tartulv.ee/datasets/c5946eb4ed264e68b17bc4145dbc9378_0/

```python
driveway = gpd.read_file("data/LI_soidutee.shp")
```

```python
driveway.info()
```

* **Liik** (class) - põhimaantee, tugimaantee, põhitänav, veotänav, jalgtee, jalgrattatee jne. (primary highway, pedestrian road etc.)
* **Tyyp** ehk eritüüp (type) - parkla tee, ringristmik, tagasipööre, tunnel jne. (roundabout, parking lot entrance etc.)
* **Era_avalik** (private/public) - riigitee, kohalik tee, eratee jne. (government road, local road etc.)
* **Suunalisus** (directionality) - 0 = kahesuunaline (two-way), 1 = ühesuunaline (one-way)

### Data preparation

#### Rename columns

```python
column_naming = {
    "OBJECTID": "object_id",
    "Number": "number",
    "Liik": "class",
    "Tyyp": "type",
    "Nimi": "name",
    "Kate": "cover",
    "Soidutee_l": "width",
    "Shape__Len": "length",
    "GlobalID": "global_id",
    "Rada_arv": "lanes",
    "Suunalisus": "directionality",
    "Era_avalik": "ownership"
}
```

```python
driveway = driveway.rename(columns=column_naming)
```

#### Set index

Setting `global_id` as index for both.

```python
driveway = driveway.set_index("global_id")
```

#### Add speed-limits

```python
speed_limits = pd.read_csv("data/piirkiirus.csv")
speed_limits = speed_limits.rename(columns={"Piirkiirus": "speed_limit", "Tee_GID": "global_id"})
speed_limits = speed_limits.set_index("global_id")
```

```python
driveway["speed_limit"] = speed_limits.speed_limit
```

```python
def add_missing_speed_limits(row):
    if np.isnan(row["speed_limit"]):
        return 50
    else:
        return row["speed_limit"]
```

```python
driveway["speed_limit"] = driveway.apply(lambda row: add_missing_speed_limits(row), axis=1)
```

```python
driveway.info()
```

#### Discard roads outside of Tartu

```python
# Load Tartu discrits 
districts = pd.read_csv("data/tartu_districts.csv", index_col=0)
# Convert to GeoDataFrame
districts['geometry'] = districts['geometry'].apply(lambda x: shapely.wkt.loads(x))
gdf_districts = gpd.GeoDataFrame(districts, geometry=districts["geometry"], crs=crs_gps).to_crs(crs_est)
# Extract city bounds
city_bounds = unary_union(gdf_districts.geometry.to_list())
```

```python
# Keep only rows that are within city limits
driveway = driveway.loc[driveway.apply(lambda row: city_bounds.contains(row.geometry), axis=1)]
```

```python
fig,ax = plt.subplots(1,1, figsize=(18,20))
# Need to change CRS for plotting
driveway.to_crs(crs_web).plot(ax=ax)
driveway.loc[["{11080A1D-1AAB-4AE4-B905-2136CCD5711F}"]].to_crs(crs_web).plot(ax=ax, color="red", lw=4)
ctx.add_basemap(ax=ax, source=ctx.providers.Stamen.TonerLite)
```

### Save as .csv

```python
driveway = driveway[["directionality", "length", "speed_limit", "geometry"]]
```

```python
driveway["directionality"] = driveway["directionality"].fillna(0)
```

```python
driveway.to_csv("data/driveways_cleaned.csv")
```
